## Task 7 – User Input with Lists

Ask the user to enter **five IP addresses**.

Store them in a list.

Then:

- Convert the list into a set
- Print the unique IP addresses
- Print the total number of unique IP addresses

```python
ip_add_list = []

for i in range(5):
  ip = input("Enter ip address: ")
  ip_add_list.append(ip)

unique_ips = set(ip_add_list)

print("unique ips")

for ip in unique_ips:
  print(ip)

print("Total Unique IP:", len(unique_ips))
```
